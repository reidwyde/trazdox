## Akaike Information Criterion

```python
model1 = growth_model.growth_model_1()
model1.sim_graph_model(model1.get_best_sim_params())
model1_aic = model1.AIC()
```

```python
model2 = growth_model.growth_model_2()
model2.sim_graph_model(model2.get_best_sim_params())
model2_aic = model2.AIC()
```

```python
plt.figure(figsize=[16,10])

plt.subplot(2,2,1)        
plt.scatter(range(6), model1_aic)
plt.title('AIC Model 1')
plt.xlabel('groups')

plt.subplot(2,2,2)
plt.scatter(range(6), model2_aic)
plt.title('AIC Model 2')
plt.xlabel('groups')


plt.subplot(2,2,3)
plt.scatter(range(6), model3_aic)
plt.title('AIC Model 3')
plt.xlabel('groups')

plt.subplot(2,2,4)
plt.scatter(range(6), model4_aic)
plt.title('AIC Model 4')
plt.xlabel('groups')

plt.show()

```
