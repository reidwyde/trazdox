# AIC and BIC




```python
import growth_model
import numpy as np
from get_tumor_db import get_tumor_db
from parse_tumor_db import parse_tumor_db
import matplotlib.pyplot as plt

```

## Log likelihood



## Akaike Information Criterion

$$ AIC = 2 \times k - 2 \times ln(\hat{L}) $$

where k = the number of parameters

## Bayesian Information Criterion

$$ BIC = \ln(n) \times k - 2 \times \ln(\hat{L}) $$

```python
model1 = growth_model.growth_model_1()
model1.sim_graph_model(model1.get_best_sim_params())
model1_aic = model1.AIC()
model1_bic = model1.BIC()
```

```python
model2 = growth_model.growth_model_2()
model2.sim_graph_model(model2.get_best_sim_params())
model2_aic = model2.AIC()
model2_bic = model2.BIC()
```

```python
model3 = growth_model.growth_model_3()
model3.sim_graph_model(model3.get_best_sim_params())
model3_aic = model3.AIC()
model3_bic = model3.BIC()

```

```python
model4 = growth_model.growth_model_4()
model4.sim_graph_model(model4.get_best_sim_params())
model4_aic = model4.AIC()
model4_bic = model4.BIC()
```

```python
plt.figure(figsize=[8,5])
plt.scatter(range(1,5), [model1_aic, model2_aic, model3_aic, model4_aic])
plt.title('AIC')
plt.xlabel('group')
plt.xticks(range(1,5))
plt.show()

plt.figure(figsize=[8,5])
plt.scatter(range(1,5), [model1_bic, model2_bic, model3_bic, model4_bic])
plt.title('BIC')
plt.xlabel('group')
plt.xticks(range(1,5))
plt.show()

print('model 1 AIC: ', model1_aic)
print('model 2 AIC: ', model2_aic)
print('model 3 AIC: ', model3_aic)
print('model 4 AIC: ', model4_aic)

print('\n\n')
print('model 1 BIC: ', model1_bic)
print('model 2 BIC: ', model2_bic)
print('model 3 BIC: ', model3_bic)
print('model 4 BIC: ', model4_bic)
```
