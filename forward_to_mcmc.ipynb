```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
# This makes the plots appear inside the notebook
%matplotlib inline
import pandas as pd
import math
import pymc3 as pm


# Initialize random number generator
np.random.seed(123)
```

```python
#forward model


def rungeKutta(ts, y0, dydt): 

    ys = np.zeros((len(ts),len(y0)))
    ys[0] = y0
    y = y0
    
    for i in range(1,(len(ts))) :
        t0 = ts[i-1]
        t = ts[i]
        h = t-t0 
        k1 = h * dydt(t0, y) 
        k2 = h * dydt(t0 + 0.5 * h, y + 0.5 * k1) 
        k3 = h * dydt(t0 + 0.5 * h, y + 0.5 * k2) 
        k4 = h * dydt(t0 + h, y + k3) 
  
        # Update next value of y 
        y = y + (1.0 / 6.0)*(k1 + 2 * k2 + 2 * k3 + k4) 
        ys[i] = y
   
    return ys



```

```python
#evolve the model

t1 = np.linspace(0, 150, 150).reshape(-1,1)
print(t1.shape)
g_orig = 9.8
sigma=1.5
noise = sigma*np.random.randn(t1.shape[0]).reshape(-1,1)
#h_orig = g_orig*t1**2 
#print(h_orig.shape)
#print(noise.shape)
#h1 = h_orig + noise
#print(h1.shape)

y0 = [0]


def dydt(t, y):
    return 9.8*t

ys = rungeKutta(t1, y0, dydt) + noise
#print(ys)

```

```python
y_exact = 0.5*g_orig*(t1**2)
y_exact = y_exact.reshape(y_exact.shape[0],1)
y_difference = ys - y_exact
plt.plot(t1, ys, t1, y_exact, "+");
y_diff = np.abs(y_difference)
```

```python

plt.semilogy(t1, y_diff)
plt.ylabel("Error")
plt.xlabel("t")
plt.title("Error in numerical integration");
# Note the logarithmic scale on the y-axis. 
```

```python

# MCMC

gravity_model = pm.Model()

with gravity_model:
    #g = pm.Uniform('g', mu = 3, sigma=5)
    
    g = pm.Uniform('g', lower=1, upper=20)
    sigma= pm.HalfNormal('sigma', sigma = 1)
    
     
    #expected outcome of variable
    mu = 1/2*g*(t1**2)
    
    #Likelihood (sampling distribution) of observations
    h_obs = pm.Normal('h_obs', mu=mu, sigma=sigma, observed=ys)
    
    
    
    
```
